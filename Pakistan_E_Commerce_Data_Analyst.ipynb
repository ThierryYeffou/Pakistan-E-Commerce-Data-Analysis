```python
#Import des bibliothèque
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Nettoyage des données**

```python
#Collection : dataset
path = "/content/data_set"
file_data = [file for file in os.listdir(path)]
#print(file_data)
for file in file_data:
  print(file)
```

```python
#Creons la base de données :
all_data = pd.DataFrame()
for file in file_data :
  all_data = pd.read_csv(path+'/'+file)
all_data
```

```python
#Observons nos données en profondeur :
#all_data.dtypes
#all_data.shape
#all_data.head()
#all_data.columns
all_data.info()
```

```python
#Comptons le nombre de valeur null
all_data.isnull().sum()
```

```python
#Suppression des valeurs manquantes
all_data = all_data.dropna(how="all")
all_data.head(5)
```

```python
all_data.isnull().sum()
```

```python
#Corrigeons l'espace sur la colonne MV :
all_data.rename(columns={' MV ':'MV'}, inplace=True)
all_data.columns
```

```python
all_data.dtypes
```

```python
#Conversion de types pour quelques colonnes
all_data['item_id']= all_data['item_id'].astype(int)
all_data['qty_ordered'] = all_data['qty_ordered'].astype(int)
all_data['Year'] = all_data['Year'].astype(int)
all_data['Month'] = all_data['Month'].astype(int)
all_data['Customer ID'] = all_data['Customer ID'].astype(str)
```

```python
#Changer la category_name_1 en catégorie
all_data.rename(columns={'category_name_1':'Category'}, inplace=True)
```

```python
all_data.columns
```

# **Analysons les 10 catégories les plus vendues et visualisons**

```python
#TopBest10Categories = all_data.groupby('Category')['item_id'].count().head(10)
TopBest10Categories = all_data.groupby('Category')['item_id'].agg(['count']).reset_index().sort_values(by='count',ascending=False).head(10)
print(TopBest10Categories)
```

```python
plt.figure(figsize=(15,5))
graph=sns.barplot(x='Category',y='count',data=TopBest10Categories)
graph.set_title('Top 10 des articles les plus vendus')
plt.show()
```
